```python
a = ['cat', 'dogs', 'frog', 'lion']
value = np.array([[40, 20, 14, 57]])
value = value.flatten()
```

```python
fig, ax = plt.subplots(figsize=(4,4), dpi = 100)
explode = [0.1, 0,0,0]
ax.pie(value, labels=a, shadow= True, explode = explode, startangle=90, autopct='%1.1f', rotatelabels=True)
plt.show()
```

```python
fig, ax = plt.subplots(dpi = 100)
value = np.array([[60,30],[73,12],[32, 87]])
value_sum = value.sum(axis=1)
value_flat = value.flatten()
size=0.3
cmap = plt.get_cmap('tab20c')
outer_colors = cmap(np.arange(3)*4)
inner_colors = cmap(np.arange(6)*3)

ax.pie(value_sum, radius=1, colors = outer_colors,
      wedgeprops= dict(width=size, edgecolor='w'))
ax.pie(value_flat, radius=1-size, colors = inner_colors,
      wedgeprops= dict(width=size, edgecolor='w'))
```

```python
fig, ax = plt.subplots(1,1, figsize=(4,4))

cmap = plt.get_cmap('tab20c')
color = cmap(np.arange(4)*3)

ax = plt.pie(value, labels=a, colors= color, autopct='%.2f')
```
